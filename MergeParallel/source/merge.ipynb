```python
import time
import random
import multiprocessing

import mergepar
```

```python
# size = int(sys.argv[-1]) if sys.argv[-1].isdigit() else 1000
size = 10000000
data_unsorted = [random.random() for _ in range(size)]
```

```python
processes = multiprocessing.cpu_count()
processes
```

```python
results = []
for i in range(1, processes+1):
    start = time.time()
    data_sorted = mergepar.merge_sort_parallel(data_unsorted, i)
    end = time.time() - start
    results.append(end)
    print(i, end, sorted(data_unsorted) == data_sorted)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, processes+1), results)
plt.xlabel('Core')
plt.ylabel('Time (s)')
plt.title('Parallel Merge sort with 1e7 elements')
plt.show()
```
